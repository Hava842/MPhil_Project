```python
import matplotlib
from sklearn import preprocessing
from scipy.stats import multivariate_normal
import numpy as np
import os
import shutil
import random

from optimizer.Optimizer import optimize

if os.path.exists('./output'):
    shutil.rmtree('./output')
os.makedirs('./output')
```

```python
def f(x):
    return np.array([(x-0.5)**2, (x+0.5)**2]) + random.uniform(-0.1,0.1)

# Deep Gauusian process. Our implementation. Not very good for few datapoints.
modelParams = {'model':'dgp',
               'maxiter': 2000,
               'layer_types': ['gp'],
               'layer_nodes': [2]}

# Gaussian Process. GPy library implementation
modelParams = {'model':'gp'}
                   
aquisitionParams = {'aquisition':'SMSego',
                    'gain': 2.0,
                    'reference': np.array([1.0, 1.0])}

frontier, curve = optimize(f,
                           np.array([[i] for i in np.linspace(-1, 1, 500)]),
                           modelParams,
                           aquisitionParams,
                           3,
                           10,
                           './output',
                           True)
```
