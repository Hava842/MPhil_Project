```python
def f(x):
    return np.array([(x-0.5)**2, (x+0.5)**2])

# Deep Gauusian process. Our implementation
# modelParams = {'model':'dgp',
#                'maxiter': 2000,
#                'layer_types': ['gp', 'noisy', 'gp', 'noisy'],
#                'layer_nodes': [2, 1, 2, 1]}

# Gaussian Process. GPy library implementation
modelParams = {'model':'gp',
               'maxiter': 2000}
                   
aquisitionParams = {'aquisition':'SMSego',
                    'gain': 2.0,
                    'reference': np.array([1.0, 1.0])}

frontier, curve = optimize(f,
                           np.array([[i] for i in np.linspace(-1, 1, 500)]),
                           modelParams,
                           aquisitionParams,
                           2,
                           10,
                           './output',
                           True)
```
